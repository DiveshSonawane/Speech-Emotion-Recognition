## Step:1 - Import Libraries

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

## Step:2 - Audio Loading & Visualization

```python
# Path to ONE sample audio file from RAVDESS (Speech)
audio_path = r"F:\Final Year Engineering\Final Year BE Project\RAVDESS\Audio_Speech_Actors_01-24\Actor_01\03-01-01-01-01-01-01.wav"
```

```python
# Load audio file
signal, sr = librosa.load(audio_path, sr=22050)
```

```python
print("Audio loaded successfully!")
print("Signal shape:", signal.shape)
print("Sampling rate:", sr)
```

```python
# Plot waveform
plt.figure(figsize=(14, 4))
librosa.display.waveshow(signal, sr=sr)
plt.title("Waveform of RAVDESS Speech Audio")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.tight_layout()
plt.show()
```

## Step:3 - MFCC Extraction & Visualization

```python
# Extract MFCC features
mfcc = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=40)
```

```python
print("MFCC shape:", mfcc.shape)
```

```python
# Plot MFCC
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfcc, x_axis='time')
plt.colorbar()
plt.title("MFCC of RAVDESS Speech Audio")
plt.xlabel("Time")
plt.ylabel("MFCC Coefficients")
plt.tight_layout()
plt.show()
```
